# Tỷ lệ gia tăng chỉ định điều trị nội trú

```python
NOI_TRU_BEFORE = BEFORE[(BEFORE['MA_LOAI_KCB'] == 3)] # Tổng số lượt nội trú KCB trong CSCKB kỳ trước
NOI_BEFORE= NOI_TRU_BEFORE.groupby(["MA_CSKCB"]).agg({'SO_LUOT':'sum'}) # Gom nhóm chi phí nội trú của các CSCKB kỳ trước
TONG_BEFORE = BEFORE.groupby(["MA_CSKCB"]).agg({'SO_LUOT':'sum'}) # gom nhóm chi phí KCB của từng CSCKB kỳ trước 

NOI_TRU_CURRENT = CURRENT[(CURRENT['MA_LOAI_KCB'] == 3)] # Lưu các chi phí nội trú kỳ hiện tại
NOI_CURRENT= NOI_TRU_CURRENT.groupby(["MA_CSKCB"]).agg({'SO_LUOT':'sum'}) # Tổng số lượt nội trú KCB trong các CSCKB kỳ hiện tại
TONG_CURRENT = CURRENT.groupby(["MA_CSKCB"]).agg({'SO_LUOT':'sum'}) # Tổng số lượt KCB của từng CSKCB kỳ hiện tại

# xử lý nối chi phí nội trú với tổng chi phí của kỳ trước
RESULT_BEFORE = pd.merge(NOI_BEFORE, TONG_BEFORE,  how='left', left_on=['MA_CSKCB'], right_on = ['MA_CSKCB'],suffixes=('_BEFORE', '_TONGLUOT')) 
# xử lý nối chi phí nội trú với tổng chi phí của kỳ hiện tại
RESULT_CURRENT = pd.merge(NOI_CURRENT, TONG_CURRENT,  how='left', left_on=['MA_CSKCB'], right_on = ['MA_CSKCB'],suffixes=('_CURRENT', '_TONGLUOT'))
# tính tỷ lệ tăng giảm của kỳ hiện tại
RESULT_CURRENT["TYLE_CHIDINHNOITRU_KYHIENTAI"] = (RESULT_CURRENT["SO_LUOT_CURRENT"] / RESULT_CURRENT["SO_LUOT_TONGLUOT"])
# tính tỷ lệ tăng giảm của kỳ trước
RESULT_BEFORE["TYLE_CHIDINHNOITRU_KYTRUOC"] = (RESULT_BEFORE["SO_LUOT_BEFORE"] / RESULT_BEFORE["SO_LUOT_TONGLUOT"])
```

```python
# xử lý nối kỳ trước và kỳ hiện tại vào 1 file
result = pd.merge(RESULT_BEFORE, RESULT_CURRENT,  how='left', left_on=['MA_CSKCB'], right_on = ['MA_CSKCB'], suffixes=('_KYTRUOC', '_KYHIENTAI'))
```

```python
# tính chi phí tăng giảm chỉ định nội trú
result["TANGGIAM_CHIDINH_DTRINOITRU"] = (result["TYLE_CHIDINHNOITRU_KYHIENTAI"] - result["TYLE_CHIDINHNOITRU_KYTRUOC"])
```

```python
# lọc các trường cần thiết khi xuất file ra Excel
total_result = result[["SO_LUOT_BEFORE","SO_LUOT_CURRENT","SO_LUOT_TONGLUOT_KYTRUOC","SO_LUOT_TONGLUOT_KYHIENTAI","TYLE_CHIDINHNOITRU_KYHIENTAI","TYLE_CHIDINHNOITRU_KYTRUOC","TANGGIAM_CHIDINH_DTRINOITRU"]].copy()
total_result.to_excel(r'D:\SOLIEU\SOLIEUPHANTICH\SOSANH\BAOGIAMGIAMSATCHIDINHNOITRU.xlsx')
```
